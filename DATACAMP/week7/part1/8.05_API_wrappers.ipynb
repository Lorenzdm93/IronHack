````python
artist_list = ['Ozzy Osbourne', 'Ariana Grande', 'Queen']
#result_list = [sp.search(q=artist, limit=50) for artist in artist_list]

def artist_to_json(artists):
    return {artist:sp.search(q=artist, limit=50) for artist in artists}

results = artist_to_json(artist_list)
````
