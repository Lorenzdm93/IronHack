```python
artistz =[]
idz=[] 

def name_id_artist(results):
    for i in results:
        try:
            artistz.append(results['tracks']['items'][0]['album']['artists'][i]['name'])
            idz.append(results['tracks']['items'][0]['album']['artists'][i]['id'])
        except:
            continue
    return artistz, idz
```

```python
def get_artist_name2(result):
    return [artists["name"] for artist in artist_list for track in result[artist]["tracks"]["items"] for artists in track["artists"]]
def get_artist_id2(result):
    return [artists["id"] for artist in artist_list for track in result[artist]["tracks"]["items"] for artists in track["artists"]]
```
