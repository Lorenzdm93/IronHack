```python
file1 = pd.read_csv('marketing_customer_analysis.csv')
```

```python
df['Response'].hist()
plt.show()
```

```python
trimmed = df.filter(items=['Response', 'Sales Channel', 'Total Claim Amount', 'Income'])
trimmed
```

```python
channelxresponse = trimmed.groupby(['Sales Channel']).count().sort_values(["Response"], ascending=True).reset_index()
channelxresponse.head()
```

```python
sns.barplot(x='Sales Channel', y='Response', data=channelxresponse)
plt.show()
```

```python
channelxresponse1 = trimmed.groupby(['Total Claim Amount']).count().sort_values(["Response"], ascending=False).reset_index()
sns.barplot(x='Total Claim Amount', y='Response', data=channelxresponse1)
plt.show()
```

```python
channelxresponse2 = trimmed.groupby(['Income']).count().sort_values(["Response"], ascending=False).reset_index()
sns.barplot(x='Income', y='Response', data=channelxresponse2)
plt.show()
```
